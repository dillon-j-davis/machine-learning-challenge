```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_period', 'koi_srad']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = df['koi_disposition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
print(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
import numpy as np
#One-hot encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
#swapping back to single digits (similar to the logistic regression model)
y_train_categorical_single=np.argmax(y_train_categorical, axis = 1)
y_test_categorical_single = np.argmax(y_test_categorical, axis = 1)
```

```python
#checking to make sure that the above code functioned properly. preview only showed values 0 and 2
for i in y_train_categorical_single:
    print(i)
```

# Train the Model



```python
#Using SVC model
from sklearn.svm import SVC 
classifier = SVC(kernel='linear')
classifier
```

```python
classifier.fit(X_train_scaled, y_train_categorical_single)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train_categorical_single)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test_categorical_single)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
             'gamma': [0.0001, 0.0005, 0.001, 0.005],
             'break_ties': [False, True]
             }
grid = GridSearchCV(classifier, param_grid, verbose = 3)
classifier.get_params().keys()
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train_categorical_single)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
